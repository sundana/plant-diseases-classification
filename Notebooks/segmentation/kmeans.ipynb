```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from k_means import k_means
import os
from tqdm import tqdm

%matplotlib inline
```

```python
IMG_PATH = "/Users/firmansyahsundana/Documents/research/computer_science/plant-diseases-classification/Data/External/New Plant Diseases Dataset(Augmented)/train"
```

```python
RESULT_PATH = "/Users/firmansyahsundana/Documents/research/computer_science/plant-diseases-classification/Notebooks/segmentation/results"
```

```python
for folder in tqdm(os.listdir(IMG_PATH)):
    os.makedirs("./results/" + folder, exist_ok=True)
    for file in os.listdir(os.path.join(IMG_PATH, folder)):
        file_path = os.path.join(IMG_PATH, folder, file)
        img = cv2.imread(file_path, cv2.COLOR_BGR2RGB)
        segmented_img = k_means(img, 2)
        cv2.imwrite(os.path.join(RESULT_PATH, folder, file + "_mask.JPG"), segmented_img)
```
